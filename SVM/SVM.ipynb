Gradient Descent:

Gradient Descent is an optimization algorithm used for minimizing the loss function in various machine learning algorithms. It is used for updating the parameters of the learning model.

w = w - α*dw

b = b - α*db

Learning Rate:

Learning rate is a tuning parameter in an optimization algorithm that determines the step size at each iteration while moving toward a minimum of a loss function.


```python
class SVM_classifier():


  # initiating the hyperparameters
  def __init__(self, learning_rate, no_of_iterations, lambda_parameter):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations
    self.lambda_parameter = lambda_parameter


  
  # fitting the dataset to SVM Classifier
  def fit(self, X, Y):

    # m  --> number of Data points --> number of rows
    # n  --> number of input features --> number of columns
    self.m, self.n = X.shape

    # initiating the weight value and bias value

    self.w = np.zeros(self.n)

    self.b = 0

    self.X = X

    self.Y = Y

    # implementing Gradient Descent algorithm for Optimization

    for i in range(self.no_of_iterations):
      self.update_weights()



  # function for updating the weight and bias value
  def update_weights(self):

    # label encoding
    y_label = np.where(self.Y <= 0, -1, 1)



    # gradients ( dw, db)
    for index, x_i in enumerate(self.X):

      condition = y_label[index] * (np.dot(x_i, self.w) - self.b) >= 1

      if (condition == True):

        dw = 2 * self.lambda_parameter * self.w
        db = 0

      else:

        dw = 2 * self.lambda_parameter * self.w - np.dot(x_i, y_label[index])
        db = y_label[index]


      self.w = self.w - self.learning_rate * dw

      self.b = self.b - self.learning_rate * db



  # predict the label for a given input value
  def predict(self, X):

    output = np.dot(X, self.w) - self.b
    
    predicted_labels = np.sign(output)

    y_hat = np.where(predicted_labels <= -1, 0, 1)

    return y_hat

    

```

```python
# Importing necessary libraries
import numpy as np  # For numerical operations, especially arrays and linear algebra

class SVM_classifier:
    """
    A custom implementation of Support Vector Machine (SVM) classifier.
    It uses gradient descent for optimization.
    """

    # Initializing hyperparameters: learning_rate, no_of_iterations, and lambda_parameter
    def __init__(self, learning_rate, no_of_iterations, lambda_parameter):
        """
        Constructor to initialize the learning rate, number of iterations,
        and lambda (regularization) parameter.
        """
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations
        self.lambda_parameter = lambda_parameter

    # Fitting the model to the dataset
    def fit(self, X, Y):
        """
        Trains the SVM model using the provided input features (X) and labels (Y).
        Implements gradient descent for optimization.
        """
        # X.shape returns (m, n) where:
        # m = number of data points (rows)
        # n = number of features (columns)
        self.m, self.n = X.shape

        # Initializing the weights (w) to zeros with size equal to the number of features
        self.w = np.zeros(self.n)

        # Initializing the bias (b) to 0
        self.b = 0

        # Storing the dataset for use in weight updates
        self.X = X
        self.Y = Y

        # Iteratively updating the weights and bias for a fixed number of iterations
        for i in range(self.no_of_iterations):
            self.update_weights()  # Call to function that updates the weights and bias

    def update_weights(self):
        """
        Updates the weights and bias using gradient descent to minimize hinge loss.
        """

        # Converting the labels: Encoding 0 as -1 (SVM works with -1, 1 labels)
        y_label = np.where(self.Y <= 0, -1, 1)

        # Looping through each data point to compute gradients and update weights
        for index, x_i in enumerate(self.X):
            # Calculating the condition to check if the point is correctly classified
            condition = y_label[index] * (np.dot(x_i, self.w) - self.b) >= 1

            if condition:
                # If correctly classified, only apply regularization (2 * λ * w)
                dw = 2 * self.lambda_parameter * self.w  # Gradient of weights
                db = 0  # No need to update bias when correctly classified
            else:
                # If incorrectly classified, apply both regularization and penalty
                dw = 2 * self.lambda_parameter * self.w - np.dot(x_i, y_label[index])
                db = y_label[index]

            # Updating weights using the gradients and learning rate
            self.w = self.w - self.learning_rate * dw

            # Updating bias using the gradient and learning rate
            self.b = self.b - self.learning_rate * db

    def predict(self, X):
        """
        Predicts the labels for the given input data points.
        """
        # Compute the linear combination of input features and weights (w·X - b)
        output = np.dot(X, self.w) - self.b

        # np.sign returns -1 for negative values and 1 for positive values
        predicted_labels = np.sign(output)

        # Converting -1 back to 0 for binary classification output
        y_hat = np.where(predicted_labels <= -1, 0, 1)

        return y_hat  # Returning the predicted labels
```
